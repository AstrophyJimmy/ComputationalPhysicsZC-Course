# $$\mathcal{Derivation \;of  \;Wien's \; Law}$$
by calculating  $\frac{d}{d\lambda}u(\lambda, T)=0$ 
where $u(\lambda, T)$ is planck's distribution formula

we reach the equation: 
$$5(e^\alpha-1)-\alpha e^\alpha=0$$
with $\alpha=\frac{hc}{\lambda K_B T}$

which to be solved numerically for $\alpha$ in the following script.

```
import numpy as np 
import matplotlib.pyplot as plt 
from mpmath import exp as e
from mpmath import *
```

```
# def func(x):
#   return 5*(e(x)-1)-x*e(x)
func = lambda x: 5*(power(e,x)-1)-x*power(e,x)
soln = float(findroot(func,4))
print('alpha = ', soln)
```

```
funcPlot = lambda x: 5*(np.exp(x)-1)-x*np.exp(x)
fig=plt.figure(figsize=(8,6))

plt.plot(np.arange(-5,5,0.01), funcPlot(np.arange(-5,5,0.01)),'--', label=r'$5(e^\alpha-1)-\alpha e^\alpha$')
plt.scatter(soln,0,s=100,alpha=0.5,c='k')
plt.text(soln,2,r'$\alpha=$'+str(np.round(soln, 5)),ha='right',fontsize=12)
plt.grid(alpha=0.2)
plt.legend(fontsize=15)
plt.show()
```
