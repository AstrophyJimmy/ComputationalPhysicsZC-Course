# Rroblem 2: Biot-Savart for a loop

```python
import numpy as np
from numpy.random import randint as rand
import matplotlib.pyplot as plt
import time
from scipy.integrate import odeint
from matplotlib.pyplot import cm
```

```python
dz = 0.1
zS = 0
zE = 1
z = np.arange(zS, zE, dz)

dtheta=2*np.pi/10
thetaS=0
thetaE=2*np.pi
theta = np.arange(thetaS, thetaE, dtheta)

B = np.zeros_like(z)
r=0.5

for i in range(z.size):
    for j in range(theta.size-1):
        Rx = -r*dtheta*np.sin(theta[j])
        Ry = r*dtheta*np.cos(theta[j])
        LX = -r*np.cos(theta[j])
        LY = -r*np.sin(theta[j])
        B[[i]] += (Rx*LY-Ry*LX)/(4*np.pi*np.power((np.sqrt(r**2+z[i]**2)),3))
plt.figure()        
plt.plot(z,B,label='Approx')
plt.plot(z,r**2/(2*np.power(r**2+np.power(z,2),3/2)), label='Exact')
plt.legend()
plt.title("BiotSavart for A loop")
plt.xlabel('z')
plt.ylabel('B')
plt.grid(alpha=0.2)
```
